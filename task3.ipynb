## Linear filtering by convolution (mean filter)

**<a id="#">Task 3.1.1.</a>** Load the image `data/astronaut.png` and show it:

```python
img = plt.imread('data/astronaut.png')
plt.imshow(img, "gray")
```

**<a id="#">Task 3.1.2.</a>** Implement a mean filter:

```python
def meanfilter(img_in, size):
    result = numpy.zeros(img_in.shape)
    n = (size - 1) // 2
    height, width = img_in.shape
    for p in numpy.ndindex(img_in.shape):
        neighborhood_sum = 0
        for q in numpy.ndindex(size, size):
            #p muss in der Mitte des "Fensters" liegen
            q_row = p[0] + (q[0]-n)
            q_coloumn = p[1] + (q[1]-n)

            if 0 <= q_row < height and 0 <= q_coloumn < width:
                neighborhood_sum += img_in[q_row, q_coloumn]
        result[p] = neighborhood_sum / (size*size)
    return result
```

```python
img_mean = meanfilter(img, 5)
plt.imshow(img_mean, "gray")
```

## Non-linear filtering

**<a id="#">Task 3.2.1.</a>** Implement a median filter:

```python
def medianfilter(img_in, size):
    result = numpy.zeros(img_in.shape)
    n = (size - 1) // 2
    height, width = img_in.shape
    for p in numpy.ndindex(img_in.shape):
        neighborhood_values = []
        for q in numpy.ndindex(size, size):
            #p muss in der Mitte des "Fensters" liegen
            q_row = p[0] + (q[0]-n)
            q_coloumn = p[1] + (q[1]-n)
            if 0 <= q_row < height and 0 <= q_coloumn < width:
                neighborhood_values.append(img_in[q_row, q_coloumn])
        neighborhood_values.sort()    
        result[p] = neighborhood_values[int(len(neighborhood_values)/2)]
    
    return result
```
