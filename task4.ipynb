**<a id="#">Task 4.1.2.</a>** Implement horizontal and vertical 3&times;3 Prewitt filters:

```python
def prewitt_v(img_in):
    n = 1
    h, w = img_in.shape
    
    result = numpy.zeros((h, w), dtype=float)
    
    mask = numpy.array([[-1, 0, 1], 
                        [-1, 0, 1], 
                        [-1, 0, 1]])
    

    for q in numpy.ndindex(3, 3):
        weight = mask[q]
        if weight == 0: continue 
        
        r_offset = q[0] - n
        c_offset = q[1] - n
        
        result[n:h-n, n:w-n] += weight * img_in[n+r_offset : h-n+r_offset, 
                                                 n+c_offset : w-n+c_offset]
    return result

def prewitt_h(img_in):
    n = 1
    h, w = img_in.shape
    result = numpy.zeros((h, w), dtype=float)
    
    mask = numpy.array([[-1, -1, -1], 
                        [ 0,  0,  0], 
                        [ 1,  1,  1]])
    
    for q in numpy.ndindex(3, 3):
        weight = mask[q]
        if weight == 0: continue
        
        r_offset = q[0] - n
        c_offset = q[1] - n
        
        result[n:h-n, n:w-n] += weight * img_in[n+r_offset : h-n+r_offset, 
                                                 n+c_offset : w-n+c_offset]
    return result
    
```

```python
plt.imshow(prewitt_h(img), "gray")

plt.imshow(prewitt_v(img), "gray")
```
